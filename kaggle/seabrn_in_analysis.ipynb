```python
from sklearn import datasets

iris = datasets.load_iris()
iris.keys()
```

```python
data = iris['data']
data.shape
```

```python
label = iris['target']
label.shape
```

```python
feature_names = iris['feature_names']
feature_names
```

```python
target_names = iris['target_names']
target_names
```

```python
import pandas as pd
import numpy as np

data_df = pd.DataFrame(data, columns=['sepal_len', 'sepal_wid', 'petal_len', 'petal_wid'])
data_df.head()
```

```python
label_df = pd.DataFrame(label, columns=['target'])
label_df.head()
```

```python
label_df['target'].value_counts()
```

```python
df = pd.concat([data_df, label_df], axis=1)
df.head()
```

```python
df.describe()
```

#  绘图

```python
import seaborn as sns
import matplotlib.pyplot as plt
f, ax = plt.subplots(1, 3, figsize=(16, 8))
sns.distplot(df[df['target'] == 0].sepal_len, ax= ax[0])
ax[0].set_title('sepal_len in 0')
sns.distplot(df[df['target'] == 1].sepal_len, ax= ax[1])
ax[1].set_title('sepal_len in 1')
sns.distplot(df[df['target'] == 2].sepal_len, ax= ax[2])
ax[2].set_title('sepal_len in 2')
plt.show()
```

### sns.pairplot(), Plot pairwise relationships in a datase

```python
# 绘制全部数值型属性的两两成对关系
sns.pairplot(df, hue='target')
plt.show()
```

```python
iris_sns = sns.load_dataset('iris')
iris_sns
```

```python
sns.pairplot(iris_sns)
```

```python
# hue 标注不同类别
sns.pairplot(iris_sns, hue='species')
plt.show()
```

```python
# 绘制指定子集的相关图
sns.pairplot(iris_sns, hue='species', vars=['sepal_length', 'sepal_width'])
```

```python
# x, y 指定不同变量
sns.pairplot(iris_sns, hue='species', x_vars = ['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
```

```python
# 修改单变量图 显示， diag_kind
sns.pairplot(iris_sns, diag_kind='kde')
```

```python
# 对散点图进行处理  kind
sns.pairplot(iris_sns, hue='species', x_vars = ['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'], kind='reg')
```
